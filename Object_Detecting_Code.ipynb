# Importing Libraries

```python
import cv2
thres = 0.5 # Threshold to detect object
```

# Importing Files

```python
classNames = []
classFile = 'coco.names'
with open(classFile, 'rt') as cf:
    classNames = cf.read().rstrip('\n').split('\n')
    
print(classNames)

configPath = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
weightsPath = 'frozen_inference_graph.pb'

net = cv2.dnn_DetectionModel(weightsPath,configPath)
net.setInputSize(320,320)
net.setInputScale(1.0/ 127.5)
net.setInputMean((127.5, 127.5, 127.5))
net.setInputSwapRB(True)
```

# Detecting Objects

```python
# img = cv2.imread('lena.png')
cam = cv2.VideoCapture(0); # '0' if using Primary Laptop Camera
cam.set(3, 1280) # Set Width
cam.set(4, 720) # Set Height

while True:
    success, img = cam.read()
    if not success:
        print("Failed to capture image from webcam.")
        continue
    
    # Perform Object Detection
    classIds, confs, bbox = net.detect(img,confThreshold=thres)
    print(classIds, bbox)
    
    if len(classIds) != 0:    
        for classIds, confidence, box in zip(classIds.flatten(), confs.flatten(), bbox):
            cv2.rectangle(img, box, color = (0,255,0), thickness = 2) # HEX RGB Color Code
            cv2.putText(img, classNames[classIds-1].upper(), (box[0]+10, box[1]+30),
                                                     cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,255), 2)
    
    # Display Output
    cv2.imshow("Output", img)
    
    # Break the loop on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close all windows
cam.release()
cv2.destroyAllWindows()
```
